```python
root = 'data/cameraman/'
img_jpeg = Image.open(root + 'cameraman.jpeg')
img = Image.open(root + 'cameraman.tiff')
img1 = Image.open(root + 'cameraman2.tiff')
img2 = Image.open(root + 'cameraman3.tiff')
img3 = Image.open(root + 'cameraman4.tiff')
img_png = Image.open(root + 'cameraman.png')
imgs = [img, img_jpeg, img_png, img1, img2, img3]
for i in imgs:
    plt.figure()
    img_ = np.array(i)
    plt.imshow(img_, 'gray')
```

```python
def info(img):
    img = np.array(img)
    print("dtype:", img.dtype)
    print("shape:", img.shape)
    print("min:", img.min(),'max', img.max())
```

```python
for i in imgs:
    info(i)
```
